```python
import requests
from bs4 import BeautifulSoup


response = requests.get("https://www.ministop.vn/vi/ms")

if response.status_code == 200:
    data = response.text 
    soup = BeautifulSoup(data, 'html.parser')

        # Find all elements with class "product-name"
    elements = soup.find_all("a")

    for el in elements:
        print(el.href)  # print text content

```
